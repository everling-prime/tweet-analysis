```python
import json
import pandas as pd
from pprint import pprint
```

```python
# Read Twitter JSON file and convert to Pandas Dataframe
with open('twitter-data.json', 'r', encoding='utf8') as f:
    line = f.readline()
    df_from_twitter_data = pd.read_json(line)
```

```python
df_from_twitter_data
```

## Home Timeline Tweets

```python
homeTimeline = df_from_twitter_data.loc['results','homeTimeline']
homeTimeline = pd.DataFrame(homeTimeline[0])
```

```python
pprint(homeTimeline.loc[1,'tweets']) # sample tweet
```

```python
### Build a dataframe with just the columns we're interested in
homeTimelineTweets = pd.DataFrame(columns=['user_id', 'text'])

for i, tweet_dict in enumerate(homeTimeline.loc[:,"tweets"]):
    if tweet_dict['text']:
        homeTimelineTweets.loc[i,'user_id'] = tweet_dict['user_id_str']
        homeTimelineTweets.loc[i,'text'] = tweet_dict['text']
        
homeTimelineTweets
```
